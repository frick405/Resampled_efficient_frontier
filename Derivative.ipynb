```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
```

```python
# 랜덤 시드, 의미 없는 데이터, 표준 정규분포
np.random.seed(8)

# 변수 설정(t: 기간, n: 기간 내 충격을 주는 횟수, x:x_0)
x = 0
t = 1
n = 1000
dt = t / n

# 데이터 프레임 구성
df = pd.DataFrame(index = np.arange(n))
list = []

# x값을 구하기 위해 n번 실행
for i in range(n):
    epsilon = float(np.random.randn())
    x = x + epsilon
    list.append(x)

df['z'] = list
plt.plot(df, color = 'g')
plt.xlabel('n')
plt.ylabel('z')
plt.title('Weiner Process')
```
